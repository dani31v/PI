NEGATIVA
-----

```python
def negativa(imagen):
    for i in range(imagen.shape[0]):
        for j in range(imagen.shape[1]):
            imagen[i,j]= 255-imagen[i,j]
```

```python
negativa(img1)
plt.imshow(img1, cmap="gray")
```

Logaritmica
----

```python
C= 255/np.log(1+imagen.max())
```

```python
img_log = C*np.log(1+imagen)
plt.imshow(img_log, "gray")
```
