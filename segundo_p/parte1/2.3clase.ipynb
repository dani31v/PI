```python
import cv2 as cv
```

```python
imagen = cv.imread("/Users/danielavalencia/Desktop/Procesamiento de ImaÃÅgenes/OpenCV-images/squirrel_cls.jpg",0)
```

```python
import matplotlib.pyplot as plt
plt.imshow(imagen)
```

```python

imagen=cv.cvtColor(imagen,cv.COLOR_GRAY2RGB)
plt.imshow(imagen)
```

```python
img_res= imagen.copy()
```

UMBRALIZACION

```python
filas,columnas=imagen.shape
```

```python


for f in range (filas):
    for c in columns:
        if imagen[f,c]>=128:
                img_res[f,c]= 255
        else:
                img_res[f,c]=0
```

```python
plt.imshow(img_res, cmap="gray")
```

OTRA MANERA DE LA UMBRALIZACI√ìN

```python
img_bin=cv.threshold(imagen,128,255, cv.THRESH_BINARY_INV)[1]
```

```python
plt.imshow(img_bin,cmap="gray")
```
