```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv.imread("../OpenCV-images/wdg2thr3.png", 0)
img_bin = cv.threshold(img, 128, 255, cv.THRESH_BINARY)[1]
```

```python
kernel_rect = cv.getStructuringElement(cv.MORPH_RECT,(5,5))
kernel_ellipse = cv.getStructuringElement(cv.MORPH_ELLIPSE,(5,5))
kernel_cross = cv.getStructuringElement(cv.MORPH_CROSS,(5,5))
```

```python
img_close = cv.morphologyEx(img_bin, cv.MORPH_CLOSE, kernel_rect, iterations=2)
plt.imshow(np.hstack((img_bin, img_close)), "gray")
```

```python
wire = cv.imread("../OpenCV-images/wire-bond.png", 0)
wire_bin = cv.threshold(wire, 128, 255, cv.THRESH_BINARY)[1]
```

```python
wire_open = cv.morphologyEx(wire_bin, cv.MORPH_OPEN, kernel_rect)
plt.imshow(np.hstack((wire_bin, wire_open)), "gray")
```
