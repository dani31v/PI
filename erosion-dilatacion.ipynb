```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv.imread('../OpenCV-images/squirrel_cls.jpg',0)
plt.imshow(img, 'gray')
```

```python
img_bin = cv.threshold(img,128,255,cv.THRESH_BINARY)[1]
```

```python
plt.imshow(img_bin, 'gray')
```

```python
SE1 = np.ones((3,3),dtype=np.uint8)
```

```python
eros = cv.erode(img_bin,SE1,iterations=4)
plt.imshow(eros, 'gray')
```
